```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
li_lists = soup.select("#top_movies_main > div > table td a")
li_lists1 = soup.select("#top_movies_main > div > table td span.tMeterScore")
li_lists2 = soup.select("#top_movies_main > div > table td.hidden-xs")
#print(li_lists1)
count  = 0

ratingLists = []
titleLists = []
countLists = []
reviewLists = []

for li in li_lists:
    count += 1
    print(count , li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
    
for l1 in li_lists1:
    print(l1.string.strip())
    ratingLists.append(l1.string.strip())
    
for l2 in li_lists2:
    print(l2.string.strip())
    reviewLists.append(l2.string.strip())     
    
```

```python
print(countLists)
print(ratingLists)
print(titleLists)
print(reviewLists)
```

```python
#countLists만으로 Dataframe 생성

df = pd.DataFrame(data=countLists, columns = ['Index'])
df['Ratings'] = pd.Series(ratingLists)
df['Titles'] = pd.Series(titleLists)
df['Reviews'] = pd.Series(reviewLists)
```

```python
df.to_csv("./Data/RottenTomatoes.csv", sep=",",encoding='utf-8',index=False)
```
